# NDVI (Normalized Difference Vegetation Index) timeseries analysis:
Analysis of a timeseries with two on-demand Sentinel-2 L2A atmospheric correction processes for Sentinel-2 (**iCor** and **smac**)

We firstly need to define range of interest in time and requested bands from Sentinel-2 L1C data

````python
center = [46.5, 11.35]
zoom = 12

eoMap = openeoMap(center,zoom)
eoMap.map
````

````python
bbox = eoMap.getBbox()
print("Coordinates selected from map:",'\n west',bbox[0],'\n east',bbox[2],'\n south',bbox[1],'\n north',bbox[3])
````

````python
bbox_poly = box(minx=bbox[0],maxx=bbox[2],miny=bbox[1],maxy=bbox[3])
print('Area Of Interest:','\nwest',bbox[0],'\neast',bbox[2],'\nsouth',bbox[1],'\nnorth',bbox[3])
spatial_extent  = {'west':bbox[0],'east':bbox[2],'south':bbox[1],'north':bbox[3]}
temporal_extent = ["2018-01-01", "2020-01-01"]

collection      = 'SENTINEL2_L1C_SENTINELHUB'
bands           = ["B04","B08","CLP",'B09','B8A','B11','sunAzimuthAngles','sunZenithAngles','viewAzimuthMean','viewZenithMean']

l1c_bands = con.load_collection(collection,spatial_extent=spatial_extent,bands=bands,temporal_extent=temporal_extent)
````

Apply atmospheric correction using iCor and smac

````python
l2a_icor = l1c_bands.atmospheric_correction(method='iCor')
````

````python
l2a_smac = l1c_bands.atmospheric_correction(method='smac')
````

Define a NDVI process where we mask out the cloudy pixels

````python
from openeo.processes import ProcessBuilder, lt
def masked_ndvi(value:ProcessBuilder):
    nir = value.array_element(1)
    red = value.array_element(0)
    clm = value.array_element(2)
    ndvi = (nir-red)/(nir+red)
    return if_(lt(clm,5),ndvi)
    
````

Take the median over the NDVI of the requested AOI and download the results.

````python
%time l2a_icor.reduce_dimension(dimension="bands",reducer = masked_ndvi ).aggregate_spatial(bbox_poly,reducer="median").download('./data/icor_ndvi_timeseries_ISDE12.nc', format="NetCDF")
````

````python
%time l2a_smac.reduce_dimension(dimension="bands",reducer = masked_ndvi ).aggregate_spatial(bbox_poly,reducer="median").download('./data/smac_ndvi_timeseries_ISDE12.nc', format="NetCDF")
````

### Plot results

````python
smac = xr.open_dataset('./data/smac_ndvi_timeseries_ISDE12.nc').dropna(dim="t").rename(band_0= "SMAC").reset_coords(['lat','lon'], drop=True)
icor = xr.open_dataset('./data/icor_ndvi_timeseries_ISDE12.nc').dropna(dim="t").rename(band_0= "iCor").reset_coords(['lat','lon'], drop=True)

data_to_plot = xr.merge([smac,icor])
data_to_plot.hvplot(x="t",dynamic=False, grid=True, width=800,height=400) * data_to_plot.hvplot.scatter(x="t")
````

# Sentinel-2 Band Sharpening
This notebook demonstrates how to perform band sharpening using the openEO process called resolution_merge based on FORCE:
https://force-eo.readthedocs.io/en/latest/components/higher-level/l2i/index.html?highlight=resolution#level-2-improphe

The process is proposed for adoption, but not yet integrated into openEO:
https://github.com/Open-EO/openeo-processes/pull/221

This notebook uses false colour 'SWIR' images to demonstrate the process.
SWIR band B12 is only available in 20m for Sentinel-2, and is combined here with B8A (20m) and B04 (10m). 
Normally, this results in an image with only 20m resolution, but the use of resolution merge will (artificially) increase the resolution.

### Retrieve 20M SWIR

````python
spatial_extent = {"west": 5.08, "south": 51.22, "east": 5.10, "north": 51.23}
bands = ["B12", "B8A", "B04"]
temporal_extent = ("2021-03-06", "2021-03-07")
collection = "TERRASCOPE_S2_TOC_V2"

cube = con.load_collection(collection,spatial_extent=spatial_extent,bands=bands,temporal_extent=temporal_extent)
````

````python
%time cube.download("./data/SWIR.nc", format="netCDF")
````

### Use the resolution_merge process and retrieve 10m SWIR

````python
%time cube.resolution_merge(high_resolution_bands=['B04'], low_resolution_bands=['B12', 'B8A']).download("./data/SWIR_resmerge.nc", format="netCDF")
````

### Retrieve 10m RGB

````python
bands=["B04","B03","B02"]

cube = con.load_collection(collection,spatial_extent=spatial_extent,bands=bands,temporal_extent=temporal_extent)
%time cube.download("./data/rgb.nc", format="netCDF")
````

### Load and plot all data 

````python
swir = xr.open_dataset('./data/SWIR.nc')
swir
````

````python
swir_data = np.moveaxis(np.array([swir.B12.values[0],swir.B8A.values[0],swir.B04.values[0]]),0,2)/10000
p = np.percentile(swir_data, 97, axis=[0,1])
swir_data = swir_data / p[ None, None,:]
````

````python
swir_hr = xr.open_dataset('./data/SWIR_resmerge.nc')
swir_data_hr = np.moveaxis(np.array([swir_hr.B12.values[0],swir_hr.B8A.values[0],swir_hr.B04.values[0]]),0,2)/10000
p = np.percentile(swir_data_hr, 97, axis=[0,1])
swir_data_hr = swir_data_hr / p[ None, None,:]
````

````python
rgb = xr.open_dataset('./data/rgb.nc')
rgb_data = np.moveaxis(np.array([rgb.B04.values[0],rgb.B03.values[0],rgb.B02.values[0]]),0,2)/10000
p = np.percentile(rgb_data, 97, axis=[0,1])
rgb_data = rgb_data / p[ None, None,:]
````

````python

fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,20))

ax1.imshow(swir_data,vmin=0,vmax=1)
ax1.set_title('SWIR_20M')

ax2.imshow(swir_data_hr,vmin=0,vmax=1)
ax2.set_title('SWIR_10M')

ax3.imshow(rgb_data,vmin=0,vmax=1)
ax3.set_title('RGB_10M')
````

## Comparison of Sigma0 backscatter computed with different processing chains: SentinelHub, SNAP

````python
W, S = 510000, 5680000
spatial_extent = {
    "west": W, "east": W + 32 * 10,
    "south": S, "north": S + 32 * 10,
    "crs": 32631
}
bands = ["VH", "VV"]
temporal_extent = ("2020-05-06T00:00:00", "2020-09-30T00:00:00")
collection = "TERRASCOPE_S2_TOC_V2"

cube = con.load_collection(collection,spatial_extent=spatial_extent,bands=bands,temporal_extent=temporal_extent)
````

````python
%%time
sar_bs = cube.sar_backscatter(coefficient="sigma0-ellipsoid")
sar_bs_dB = sar_bs.apply(lambda x: 10 * x.log(base=10))
sar_bs_dB.download("../data/US3/shub-series.nc", format="NetCDF")
````

````python
%%time
# Ascending orbit data:
collection = "S1_GRD_SIGMA0_ASCENDING"
asc = con.load_collection(collection,spatial_extent=spatial_extent,bands=bands,temporal_extent=temporal_extent
# Descending orbit data:
collection = "S1_GRD_SIGMA0_DESCENDING"
desc = con.load_collection(collection,spatial_extent=spatial_extent,bands=bands,temporal_extent=temporal_extent
# Datacubes merging:
desc.merge_cubes(asc,overlap_resolver="max").apply(lambda x: 10 * x.log(base=10)).download("../data/US3/snap-series.nc", format="NetCDF") 
````

````python
shub_ts = xr.open_dataset("../data/US3/shub-series.nc").rename({"VV": "VV_SHUB", "VH":"VH_SHUB"})
snap_ts = xr.open_dataset("../data/US3/snap-series.nc").rename({"VV": "VV_SNAP", "VH":"VH_SNAP"})
````

````python
data_to_plot = xr.merge([shub_ts, snap_ts]).mean(dim=['x','y'])
data_to_plot.hvplot(dynamic=False,width=850,height=600, grid=True, alpha=0.3, ylabel = 'dB') * data_to_plot.hvplot.scatter()
````
